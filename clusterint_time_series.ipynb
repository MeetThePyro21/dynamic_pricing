```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()

%matplotlib inline
```

```python
df_transactions = pd.read_csv('transactions.csv')
df_transactions = df_transactions[['dates', 'SKU', 'price']]
df_transactions.dates = pd.to_datetime(df_transactions.dates)
```

```python
pivot_t = df_transactions.pivot_table(index='dates',columns='SKU', values='price')
pivot_t
```

```python
pd.DataFrame(pivot_t.to_records())
```

```python
for col in pivot_t.columns:
    pivot_t.loc[:,col] = pivot_t.loc[:,col].ffill()
    
for col in pivot_t.columns:
    pivot_t.loc[:,col] = pivot_t.loc[:,col].bfill()
    
pivot_t.reindex(columns=[i for i in pivot_t.columns]) 
```

```python
to_clustering = pivot_t.loc[pd.to_datetime('2019-01-01'):pd.to_datetime('2019-11-30')]
```

```python
to_clustering.columns = [i for i in to_clustering.columns]
```

```python
from tslearn.clustering import TimeSeriesKMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
scaler = StandardScaler()
ts_scaled = scaler.fit_transform(to_clustering).T
```

```python
ts_scaled.shape
```

```python
from tqdm.notebook import tqdm
```

```python
distortions = []
silhouette = []
K = range(1, 10)
for k in tqdm(K):
    kmeanModel = TimeSeriesKMeans(n_clusters=k, metric="euclidean", n_jobs=6, max_iter=50)
    kmeanModel.fit(ts_scaled)
    distortions.append(kmeanModel.inertia_)
    if k > 1:
        silhouette.append(silhouette_score(ts_scaled, kmeanModel.labels_))

plt.figure(figsize=(10,4))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('Elbow Method')
plt.show()

plt.figure(figsize=(10,4))
plt.plot(K[1:], silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette score')
plt.title('Silhouette')
plt.show()
```

```python
n_clusters = 5
ts_kmeans = TimeSeriesKMeans(n_clusters=n_clusters, metric="euclidean", n_jobs=3, max_iter=10)
ts_kmeans.fit(ts_scaled)
```

```python
ts_kmeans.predict(ts_scaled)
```

```python
cl_d = dict()
for cluster, sku in zip(ts_kmeans.predict(ts_scaled),to_clustering.columns):
    if cl_d.get(str(cluster)) is None:
        cl_d[str(cluster)] = [sku]
    else:
        cl_d[str(cluster)] = cl_d[str(cluster)] + [sku]
```

```python
for cluster in cl_d.keys():
    
    plt.figure(figsize=(12,8))
    sns.lineplot(data = pivot_t[cl_d[cluster][:20]], dashes=False)
```
