```python
import pandas as pd
import numpy as np

#import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()

%matplotlib inline
```

```python
df = pd.read_csv('/Users/awesome/Downloads/test.csv')
df.dates = pd.to_datetime(df.dates)
df.head()
```

```python
# Imputing with KNNImputer
from sklearn.impute import KNNImputer

first_group_sku = pd.read_csv('/Users/awesome/Downloads/first_group_sku.csv')
second_group_sku = pd.read_csv('/Users/awesome/Downloads/second_group_sku.csv')
third_group_sku = pd.read_csv('/Users/awesome/Downloads/third_group_sku.csv')

# Define a subset of the dataset
df_knn_1 = first_group_sku.filter(
    ['SKU', 'price_per_sku', 'num_purchases'], axis=1).copy()
df_knn_2 = second_group_sku.filter(
    ['SKU', 'price_per_sku', 'num_purchases'], axis=1).copy()
df_knn_3 = third_group_sku.filter(
    ['SKU', 'price_per_sku', 'num_purchases'], axis=1).copy()


# Define KNN imputer and fill missing values
knn_imputer_1 = KNNImputer(
    n_neighbors=10, weights='uniform', metric='nan_euclidean')
df_knn_imputed_1 = pd.DataFrame(
    knn_imputer_1.fit_transform(df_knn_1), columns=df_knn_1.columns)

knn_imputer_2 = KNNImputer(
    n_neighbors=10, weights='uniform', metric='nan_euclidean')
df_knn_imputed_2 = pd.DataFrame(
    knn_imputer_2.fit_transform(df_knn_2), columns=df_knn_2.columns)

knn_imputer_3 = KNNImputer(
    n_neighbors=10, weights='uniform', metric='nan_euclidean')
df_knn_imputed_3 = pd.DataFrame(
    knn_imputer_3.fit_transform(df_knn_3), columns=df_knn_3.columns)

df_knn = pd.concat([df_knn_imputed_1, df_knn_imputed_2,
                    df_knn_imputed_3]).reset_index(drop=True)

df_group = pd.concat([first_group_sku, second_group_sku,
                     third_group_sku]).reset_index(drop=True)

df_group['price_per_sku'] = df_knn['price_per_sku'].values
df_group.dates = pd.to_datetime(df_group.dates)
df_group.head()
```

```python
df_group['year'] = df_group.dates.dt.year
df_group['month'] = df_group.dates.dt.month
df_group['quarter_of_year'] = df_group.dates.dt.quarter
df_group['week_of_year'] = df_group.dates.dt.weekofyear
df_group['day_of_year'] = df_group.dates.dt.dayofyear
df_group['day_of_month'] = df_group.dates.dt.day
df_group['day_of_week'] = df_group.dates.dt.dayofweek
df_group["is_wknd"] = df_group.dates.dt.weekday // 4
df_group['is_month_start'] = df_group.dates.dt.is_month_start.astype(int)
df_group['is_month_end'] = df_group.dates.dt.is_month_end.astype(int)
df_group = df_group.drop(columns=['dates'])
df_group.head()
```

```python
df['year'] = df.dates.dt.year
df['month'] = df.dates.dt.month
df['quarter_of_year'] = df.dates.dt.quarter
df['week_of_year'] = df.dates.dt.weekofyear
df['day_of_year'] = df.dates.dt.dayofyear
df['day_of_month'] = df.dates.dt.day
df['day_of_week'] = df.dates.dt.dayofweek
df["is_wknd"] = df.dates.dt.weekday // 4
df['is_month_start'] = df.dates.dt.is_month_start.astype(int)
df['is_month_end'] = df.dates.dt.is_month_end.astype(int)
df = df.drop(columns=['dates'])
df.head()
```

```python
train_df = df_group
test_df = df
```

```python
Y_train = train_df['price_per_sku']
X_train = train_df.drop(columns=['price_per_sku', 'num_purchases'])

X_test = test_df
```

```python
import lightgbm as lgb

params = {
    'n_estimators': 1000,
    'max_depth': 5,
#     'num_leaves': 2**4,
    'learning_rate': 0.1,
#     'boosting_type': 'dart'
}
```

```python
model = lgb.LGBMRegressor(first_metric_only = True, **params)

model.fit(X_train, Y_train,
#           early_stopping_rounds = 10,
          verbose = 0)
```

```python
forecast = model.predict(X_test)
```
